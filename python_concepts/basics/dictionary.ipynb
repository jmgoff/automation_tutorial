dictionaries

 similar to lists, but keys are manually assigned to elements (a.k.a. values)

```python
d = {1:2,
'c':3,
3:'d'
}
```

```python
print (d[1])
print (d['c'])
print (d[3])
```

keys: items used to access elements

```python
print(d.keys())
# values: elements corresponding to respective keys
print(d.values())
```
