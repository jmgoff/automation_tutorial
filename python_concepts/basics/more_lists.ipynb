accessing items in python lists<br>
 (and other list-like types)

```python
a = [1,2,3]
```

items in a python list are assigned base-zero indices

```python
print(a[0])
print(a[1])
print(a[2])
```

values may be accessed and modified index-wise

```python
a[0] = 6
```

```python
print (a)
```

you can insert other items into a list<br>
 append new element at the end of the list

```python
a.append(3) 
print(a)
```

 add new element at a specified index:

```python
a.insert(0,1)
print(a)
## syntax is x.insert(<index>,item)
```

the index corresponding to the first appearance of an element may be found<br>
 with the '.index' function

```python
print(a.index(6))
```
